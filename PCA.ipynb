# Clustering with Original Data:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
wine=pd.read_csv('/content/wine.csv')
wine
```

```python
# Normalization function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_wine_df = scaler.fit_transform(wine.iloc[:,0:])
scaled_wine_df
```

```python
#WCSS (Within Cluster Sum of Squares)
wcss = []
for i in range(1, 11):    #Number of clusters
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_wine_df)
    wcss.append(kmeans.inertia_)

print(wcss)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=0)
model.fit(scaled_wine_df)
```

```python
model.labels_
```

```python
#Assign clusters to the data set
wine['ClusterID'] = model.labels_
wine
```

```python
print(wine[wine['ClusterID']==0])
print(wine[wine['ClusterID']==1])
print(wine[wine['ClusterID']==2])
```

```python
#These are standardized values.
model.cluster_centers_
```

```python
#wine.groupby('clusterid_new').agg(['mean']).reset_index()
wine.iloc[:,1:].groupby('ClusterID').agg(['mean'])
#wine.groupby('clusterid_new').mean()
```

```python
from sklearn import metrics
for i in range(2,5):
  model = KMeans(n_clusters=i, random_state=0)
  model.fit(scaled_wine_df)
  label=model.labels_
  wine['N_' + str(i)]= model.labels_
  print(label)
  print('The Silhouette Score for',i,'clusters is',metrics.silhouette_score(scaled_wine_df, label))
#metrics.silhouette_score(scaled_wine_df, model.labels_)
wine
```

##2. Discuss Similarities and Differences in Clustering Results
##Similarities:

- All clustering solutions (2, 3, and 4 clusters) resulted in distinct groupings of data points.

- Higher clusters show variations among the groupings and demonstrate distinct patterns.

##Differences:

- Silhouette Scores: The 2-cluster solution had a moderate Silhouette Score (0.2862), the 3-cluster solution had the highest score (0.3065), and the 4-cluster solution had the lowest score (0.2505). This indicates that the 3-cluster solution provided the best separation and defined clusters.

- Cluster Assignments: The cluster assignments changed significantly when increasing the number of clusters. The 3-cluster solution provided a clear distinction, whereas the 4-cluster solution introduced more complexity without better separation.

---------------------------------------------------------------------

##3. Impact of Dimensionality Reduction on Clustering Performance
###Dimensionality reduction using PCA can have significant impacts on clustering performance:

- Simplification: PCA reduced the dimensionality, simplifying the dataset while preserving the majority of the variance (67.71% with three principal components). This can make clustering more efficient and reduce noise.

- Improved Clarity: Clustering on PCA-transformed data can enhance the clarity and separation of clusters. The cumulative variance explained by PCA helped in identifying the principal components, which captured the essential patterns in the data.

- Performance: The higher Silhouette Score for the 3-cluster solution indicates that dimensionality reduction improved the clustering performance by defining clearer and more distinct clusters.
-------------------------------------------------------------------

##4. Trade-offs Between Using PCA and Clustering Directly on the Original Dataset
###Using PCA:

###Pros:

- Reduction in Complexity: PCA reduces the number of features, making the dataset more manageable and less prone to overfitting.

- Enhanced Clustering: By focusing on the principal components that capture the most variance, PCA can improve the quality and separation of clusters.

- Noise Reduction: PCA can filter out noise and irrelevant features, leading to more robust clustering.

###Cons:

- Loss of Information: While PCA retains the majority of the variance, some information is inevitably lost in the dimensionality reduction process.

- Interpretability: The transformed features may be less interpretable compared to the original features, making it harder to understand the meaning behind each principal component.
---------------------------------------------------------------------------

#1. Key Findings and Insights
Principal Component Analysis (PCA): PCA effectively reduced the dataset's dimensions while retaining 67.71% of the variance with three principal components. This reduction simplified the data and highlighted the most significant patterns.

- Clustering Analysis: The clustering solutions were evaluated for 2, 3, and 4 clusters, with the 3-cluster solution achieving the highest Silhouette Score (0.3065), indicating the best cluster separation and definition.

- Comparison of Results: The combination of PCA and clustering provided more clearly defined clusters and better separation than clustering directly on the high-dimensional dataset.

#2. Practical Implications of Using PCA and Clustering in Data Analysis
- Efficiency: PCA reduces the dimensionality of the dataset, making clustering and other analyses more efficient by lowering computational complexity and processing time.

- Noise Reduction: PCA helps filter out noise and irrelevant features, leading to more robust and accurate clustering results.

- Interpretability: Clustering on PCA-transformed data can enhance the clarity and separation of clusters. However, the transformed features might be less interpretable compared to the original features, making it challenging to understand the underlying meaning of each principal component.

- Data Visualization: PCA can facilitate the visualization of high-dimensional data in lower dimensions, making it easier to identify patterns and relationships within the dataset.

#3. Recommendations for When to Use Each Technique
##Use PCA Before Clustering:

- When dealing with high-dimensional datasets to reduce complexity and highlight the most significant patterns.

- When aiming to improve clustering performance by eliminating noise and irrelevant features.

- When computational efficiency is crucial, as PCA can reduce the time and resources needed for clustering.

##Clustering Directly on the Original Dataset:

- When the dataset has a manageable number of features, and dimensionality reduction is not necessary.

- When interpretability of the original features is essential for understanding the clusters.

- When retaining all available information is crucial for the analysis, and no significant noise is present.

#Conclusion
####The choice between using PCA and clustering directly on the original dataset depends on the specific goals of the analysis and the nature of the data. PCA is beneficial for simplifying high-dimensional data, reducing noise, and improving clustering performance, while clustering on the original dataset provides more interpretable and detailed results.
